```python
import pandas as pd
```

```python
df=pd.read_csv('/content/car_dekho_cleaned_dataset_Raw.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df['it']=df['it'].astype(str)
df['ownerNo']=df['ownerNo'].astype(str)
df['modelYear']=df['modelYear'].astype(str)
df['Seats']=df['Seats'].astype(str)
```

```python
X=df.drop(columns='price')
y=df['price']
```

```python
import scipy.stats as stats
categorical_columns = ['it', 'ft', 'bt', 'modelYear', 'transmission',  'oem', 'City', 'Seats']

anova_results = {}
for col in categorical_columns:
    groups = [df[df[col] == level]['price'] for level in df[col].unique()]
    if len(groups) > 1:  # Ensure there are at least two groups
        f_stat, p_value = stats.f_oneway(*groups)
        anova_results[col] = {'F-statistic': f_stat, 'P-value': p_value}
```

```python
anova_results
```

```python
numerical_columns = ['mileage', 'km']
anova_result = {}
for col in numerical_columns:
    groups = [df[df[col] == level]['price'] for level in df[col].unique()]
    if len(groups) > 1:  # Ensure there are at least two groups
        f_stat, p_value = stats.f_oneway(*groups)
        anova_result[col] = {'F-statistic': f_stat, 'P-value': p_value}
```
